```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
iris_data = pd.DataFrame(iris.data)
iris_data.columns = iris.feature_names
iris_data['Type']=iris.target
iris_data.head()
print(iris_data.all)
print("Kurtosis result : {}".format(iris_data.kurtosis()))
print("Skew result : {}".format(iris_data.skew()))
```

```python
from scipy.cluster.vq import kmeans, whiten, kmeans2
from scipy import cluster
import numpy as np

cluster_number = 3
column = 'sepal width (cm)'
column2 = 'sepal length (cm)'

axe1 = iris_data[column]
axe2 = iris_data[column2]
iris_X = iris_data[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
# Get centroids and average distance
#centroids, avg_distance = kmeans(axe1, cluster_number)
# Get the groups (clusters) and distances
#groups, cluster_distance = cluster.vq.vq(axe1, centroids)

# Get centroids and average distance
centroids, avg_distance = kmeans2(iris_X, cluster_number)
# Get the groups (clusters) and distances
groups, cluster_distance = cluster.vq.vq(iris_X, centroids)

print(centroids)
print(avg_distance)
print(groups)
print(cluster_distance)
kmeans_result = pd.DataFrame(
    {'Column N°{}'.format(column): axe1,
     'Distance to centroid': cluster_distance,
     'Column N°{}'.format(column2): axe2,
     'Cluster': groups
     })
print(kmeans_result.all)
```

```python
axe1 = axe1.reshape(-1, 1)
clus = plt.scatter(axe2, axe1, s=cluster_distance*100, c=groups)#, edgecolors='r', facecolors='none')
plt.ylabel('Column N°{}'.format(column))
plt.xlabel('Column N°{}'.format(column2))
plt.show(clus)
```
