```python
def convert(filename):
    print("*********************"+filename+"************************")
    mat = scipy.io.loadmat(filename+".mat")
    print("**********mat version***********")
    print(mat)
    data = pd.DataFrame(np.hstack((mat['X'], mat['y'])))
    data.to_csv("results/"+filename+".csv")
    print("**********dataframe version***********")
    print(data)
```
