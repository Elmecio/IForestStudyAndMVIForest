```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#import sys
import datasets as datat
dts = datat.synthetic_datasets()
util = datat.utilitaries()
```

```python
def show_figure(full_dataset, dataset_name) :
    # Load the dataset
    X_normal,Y_normal, X_abnormal, Y_abnormal = util.normals_abnormals_XY(full_dataset, -1)
    # Plot data
    plt.scatter(X_abnormal[0], X_abnormal[1], c="r")
    plt.scatter(X_normal[0], X_normal[1], c="b")
    plt.xlim(-30,30)
    plt.ylim(-20,20)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(dataset_name)
    # Save picture
    plt.savefig("synthetic_2D_data/Synthetics_2D/useddatasets/"+dataset_name+".png", bbox_inches='tight', pad_inches=0.05)
    
    plt.show()
```

```python
X_brut, y_transform, full_dataset, x_lim, y_lim, z_lim, data_Version  = dts.TwoD2_V3_3_0()
dataset_name = "Synthetic 1"
show_figure(full_dataset, dataset_name)
```

```python
X_brut, y_transform, full_dataset, x_lim, y_lim, z_lim, data_Version  = dts.TwoD2_V3_4_0()
dataset_name = "Synthetic 2"
show_figure(full_dataset, dataset_name)
```
