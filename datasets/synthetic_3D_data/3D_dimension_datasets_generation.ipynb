# 3-Dimensions synthetics datasets generation

```python
%matplotlib inline
#%matplotlib notebook
#calling it a second time may prevent some graphics errors
#%matplotlib notebook  

import sys
sys.path.append('../')
import datasets_generator as dg
data_gene = dg.synthetic_datasets(path_top="../")
```

## Generals parameters

```python
n_dimension = 3
n_size = 1500 #number of normal samples
contamination = 1/100 # 1%
size_a = int(n_size*contamination) #size of abnormal samples
```

## Version 1.0.0

```python
version = "1.0.0"

#Normal data
n_min_circle = 2 #Rayon
n_max_circle = 7 #Rayon
n_distribution="uniform"

#Abnormal data
a_max_circle = 2 #Rayon
a_min_circle = 0.0 #Rayon
a_distribution="uniform"

x_lim = 10
y_lim = 10
z_lim = 7

min_value = -10
max_value = 10
mu = 0.0
sigma = 0.0

data_gene.dataset_generator(n_size=n_size, n_small_circle=n_min_circle,n_big_circle=n_max_circle, 
                            n_distribution=n_distribution, 
                            a_size=size_a, a_small_circle=a_min_circle,a_big_circle=a_max_circle, 
                            a_distribution=a_distribution,
                            x_lim=x_lim, y_lim=y_lim,z_lim=z_lim, min_value=min_value, max_value=max_value,
                            mu=mu, sigma=sigma, version=version,n_dimension = n_dimension)
```
