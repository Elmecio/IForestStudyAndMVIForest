```python
data_path = datat.path_HTTP_IForestASD
dts = datat.reals_datasets(path_top="../")
X_full_brut, y_full_transform, full_dataset = dts.HTTP_IForestASD()
print("X_full_brut")
print(X_full_brut.describe())
print("y_full_transform")
print(y_full_transform.describe())
print("full_dataset")
print(full_dataset.describe())
print("***************************************************************************************")
X_normals_brut, y_normals_transform, full_normals_dataset = datat.only_normals(data_path, '\'Anomaly\'')
print("X_normals_brut")
print(X_normals_brut.describe())
print("y_normals_transform")
print(y_normals_transform.describe())
print("full_normals_dataset")
print(full_normals_dataset.describe())
print("***************************************************************************************")
X_abnormals_brut, y_abnormals_transform, full_abnormals_dataset = datat.only_abnormals(data_path, '\'Anomaly\'')
print("X_abnormals_brut")
print(X_abnormals_brut.describe())
print("y_abnormals_transform")
print(y_abnormals_transform.describe())
print("full_abnormals_dataset")
print(full_abnormals_dataset.describe())
print("% of anomaly in the dataset")
print(str(len(full_abnormals_dataset)*100/len(full_dataset))+"%")
```
