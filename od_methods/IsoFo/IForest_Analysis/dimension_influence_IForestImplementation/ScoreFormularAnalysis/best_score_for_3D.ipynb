## Test with 3 dimension dataset

```python
import sys
sys.path.append('../../../../../')
from datasets import datasets as datat

from metrics import threshold_evaluator
t_evaluator = threshold_evaluator.functions_threshold_evaluator()
```

## Generals settings

```python
dts = datat.synthetic_datasets(path_top="../../../../../")
max_samples = 256
n_trees = 100
threshold = 0.6
min_sample_size = 5
sample_size_step = 5
```

## Executions

### Dataset ThreeD_V1_0
#### Conclusion on this dataset :
IForest better than EIF

```python
# Load the dataset
X_brut, y_transform, full_dataset = dts.ThreeD_V1_0()
x_lim = 10
y_lim = 7
#X_brut.describe()
#pd.set_option("display.max_rows", None, "display.max_columns", None)
#print(X_brut)
print("################################################## IFOREST ###################################")
t_evaluator.execute_IForest(X_brut=X_brut, max_samples=max_samples, n_trees=n_trees, threshold=threshold, 
                y_transform=y_transform, x_lim=x_lim, y_lim=y_lim)
print("################################################## EIF ###################################")
t_evaluator.execute_eif(X_brut=X_brut, max_samples=max_samples, n_trees=n_trees, threshold=threshold, 
                y_transform=y_transform, x_lim=x_lim, y_lim=y_lim)
```
