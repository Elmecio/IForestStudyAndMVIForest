## Generals settings

```python
max_samples = 256
n_trees = 100
contamination = 0.1
threshold = 0.6
```

## Chargement du jeu de donn√©es

```python
dts = datat.synthetic_datasets()
#X_brut, y_transform = dts.OneD_V1_1()
X_brut, y_transform, full_dataset = datat.only_normals('../../../../'+dts.path_1D_V1_1)
X_brut.describe()
```

## Execution of IForest on the dataset

```python
specificities =[]
roc_aucs = []
for i in range(5):
    # Instanciation, fit and predict on Shuttle
    func_IFD = IFD.IsolationForest(sample_size = max_samples, n_trees=n_trees)
    IFD_y_pred_IF, scores = func_IFD.fit_predict(X=X_brut, threshold=threshold)
    
    print(confusion_matrix(y_transform, IFD_y_pred_IF))
    
    print("Score min = "+str(np.min(scores)))
    print("Score max = "+str(np.max(scores)))
    visu.plot_2D(X_brut[0], scores,'Data', 'Scores', "Anomaly scores").show()
```
