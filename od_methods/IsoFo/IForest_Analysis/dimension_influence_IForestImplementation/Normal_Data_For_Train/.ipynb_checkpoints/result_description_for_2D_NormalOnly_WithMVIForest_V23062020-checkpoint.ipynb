# Result description of IForest, MVIForest, EIF on synthetic datasets. The trainning phase only on normal data

## Test with 2 dimension dataset

```python
import sys
sys.path.append('../../../../../')

from datasets import datasets as datat
dts = datat.synthetic_datasets(path_top="../../../../../")
util = datat.utilitaries()
from metrics import useful_infos as ue

from experiments import trainning_parametors
t_parametors = trainning_parametors.functions_train_parametors()
```

## Generals settings

```python
dts = datat.synthetic_datasets(path_top="../../../../../")
max_samples = 256
n_trees = 100
threshold = 0.6
min_sample_size = 5
sample_size_step = 5
save_fig=True
```

## Executions

### Dataset TwoD_V1_0
#### Conclusion on this dataset :
IForest better than EIF

```python
# Load the dataset
X_brut, y_transform, full_dataset = dts.TwoD_V1_0()
X_normal,Y_normal, X_abnormal, Y_abnormal = util.normals_abnormals_XY(
    dataset=full_dataset, outlier_label=ue._OUTLIER_PREDICTION_LABEL)
dataset_name = "NormalOnly_TwoD_V1_0_0"
x_lim = 10
y_lim = 7
#X_brut.describe()
#pd.set_option("display.max_rows", None, "display.max_columns", None)
#print(X_brut)
print("################################################## IFOREST ###################################")
t_parametors.execute_IForest(X_brut=X_brut, max_samples=max_samples, n_trees=n_trees, threshold=threshold, 
                y_transform=y_transform, x_lim=x_lim, y_lim=y_lim, dataset_name=dataset_name, save_fig=save_fig, 
                             X_train=X_normal, show_samples=True)
print("################################################## MVIFOREST ###################################")
t_parametors.execute_MVIForest(X_brut=X_brut, max_samples=max_samples, n_trees=n_trees, threshold=threshold, 
                y_transform=y_transform, x_lim=x_lim, y_lim=y_lim, dataset_name=dataset_name, save_fig=save_fig, 
                             X_train=X_normal, show_samples=True)
print("################################################## EIF ###################################")
t_parametors.execute_eif(X_brut=X_brut, max_samples=max_samples, n_trees=n_trees, threshold=threshold, 
                y_transform=y_transform, x_lim=x_lim, y_lim=y_lim, dataset_name=dataset_name, save_fig=save_fig, 
                             X_train=X_normal, show_samples=True)
```
