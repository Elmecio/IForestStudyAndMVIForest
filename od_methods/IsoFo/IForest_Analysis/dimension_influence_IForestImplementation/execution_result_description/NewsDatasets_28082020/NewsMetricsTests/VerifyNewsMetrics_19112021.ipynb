```python
def show_figure(data_path, dataset_name, method_name) :
    # Load the dataset
    dataset = datat.load_data_from_CSV(data_path);
    # Split in execution results
    XBrut, YBrut, YClassified, Scores, PathLengths = util.split_in_results(dataset)
    
    #print(YBrut) 
    #print(YClassified)
    print("Confusion matrice details")
    print("    tn  fp")
    print("    fn  tp")
    print("          ")
    tn, fp, fn, tp, cm, auc, spec, prec, rec, f1, far = perf.performance_summary(Y_predict=YClassified, Y_original=YBrut, 
                                                                                     scores=Scores, CPUTime = None, 
                                                                                     UseMemory=None, print_result=True, 
                                                                                     outputPRAUC = False, labels = [1,-1],
                                                                                     pos_label=-1)
    print("tn = "+str(tn)+", fp = "+str(fp)+", fn = "+str(fn)+", tp = "+str(tp))
    print("Specificity : " + str(tn/(tn+fp)))
    print("Precision : " + str(tp/(tp+fp)))
    print("Recall : " + str(tp/(tp+fn)))
    print("f1_score : " + str(2*((tp/(tp+fp) * tp/(tp+fn))/(tp/(tp+fp) + tp/(tp+fn)))))
    print("False alarm rate (%) : " + str(fp/(fp+tn)))
    
    print("######################################################################################################")
    
    
    #YBrut = YBrut[2].replace(1, 0)
    #YBrut = pd.DataFrame(YBrut)
    #YBrut = YBrut[2].replace(-1, 1)
    
    #YClassified = YClassified[3].replace(1, 0)
    #YClassified = pd.DataFrame(YClassified)
    #YClassified = YClassified[3].replace(-1, 1)
    #print(YBrut) 
    #print(YClassified)
    #print("Confusion matrice details")
    #print("    tp  fn")
    #print("    fp  tn")
    #print("          ")
    tn, fp, fn, tp, cm, auc, spec, prec, rec, f1, far = perf.performance_summary(Y_predict=YClassified, Y_original=YBrut, 
                                                                                     scores=Scores, CPUTime = None, 
                                                                                     UseMemory=None, print_result=True, 
                                                                                     outputPRAUC = False, labels = [-1,1],
                                                                                    pos_label=1)
    print("tn = "+str(tn)+", fp = "+str(fp)+", fn = "+str(fn)+", tp = "+str(tp))
    print("Specificity : " + str(tn/(tn+fp)))
    print("Precision positive : " + str(tp/(tp+fp)))
    print("Precision negative : " + str(tn/(tn+fn)))
    print("Recall : " + str(tp/(tp+fn)))
    print("f1_score positive : " + str(2*((tp/(tp+fp) * tp/(tp+fn))/(tp/(tp+fp) + tp/(tp+fn)))))
    print("f1_score negative : " + str(2*((tn/(tn+fn) * tp/(tp+fn))/(tn/(tn+fn) + tp/(tp+fn)))))
    print("False alarm rate (%) : " + str(fp/(fp+tn)))
```
