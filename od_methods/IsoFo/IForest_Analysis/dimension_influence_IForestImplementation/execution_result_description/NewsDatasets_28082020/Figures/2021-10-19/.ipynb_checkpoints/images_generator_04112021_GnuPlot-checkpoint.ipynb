```python
from pygnuplot import gnuplot
import subprocess
import pandas as pd
```

### ImpactDeTailleEchantillon_IForest_Shuttle_2020-10-04-21-57-06-686354.csv

```python
dataset = pd.read_csv("ImpactDeTailleEchantillon_IForest_Shuttle_2020-10-04-21-57-06-686354.csv", header=None, index_col=None)
print(dataset)
dataset.drop([0], inplace=True)
print(dataset)
```

```python
g = gnuplot.Gnuplot(output = '"recall_specificity.ps"', key = 'right center')
g.set('ter pos "Helvetica"16')
g.set(xlabel ='"Sample Size"', ylabel = '"Recall"', yrange='[0.82:1.01]')
g.set('ytics nomirror')
g.set(y2label='"Specificity"', y2range='[0.1:1.01]')
g.set('ytics')
g.set('y2tics')
g.set('xtics')
g.plot_data(dataset, 'using 2:5 axes x1y1 with lines title "Recall"',
                       'using 2:6 axes x1y2 with lines title "Specificity"')

#Convert to pdf
subprocess.call(["ps2pdf", "recall_specificity.ps", "recall_specificity.pdf"])

#from IPython.display import IFrame
#IFrame("recall_specificity.pdf", width=1000, height=1000)
```

```python
g = gnuplot.Gnuplot(output = '"roc_auc.ps"')
g.set('ter pos "Helvetica"16')
g.set(xlabel ='"Sample Size"', ylabel = '"ROC AUC"', yrange='[0.99:1]')
g.set('xtics')
g.unset('key')
g.plot_data(dataset, 'using 2:4 with lines')

#Convert to pdf
subprocess.call(["ps2pdf", "roc_auc.ps", "roc_auc.pdf"])

#from IPython.display import IFrame
#IFrame("recall_specificity.pdf", width=1000, height=1000)
```
