```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Create some mock data

dataset = pd.read_csv("ImpactDeTailleEchantillon_IForest_Shuttle_2020-10-04-21-57-06-686354.csv", header=None, index_col=None)
dataset.drop([0], inplace=True)
#print(dataset)

sampleSize = dataset[0]
#print(sampleSize)
ROCAUC = dataset[2].astype(float)
#print(ROCAUC)
Recall = dataset[3].astype(float)
#print(Recall)
Specificity = dataset[4].astype(float)
#print(Specificity)
#df.round(1)


fig, ax1 = plt.subplots()

ax1.set_xlabel('Sample size')
sampleSize = range(5000)

ax1.set_ylabel('Recall')
ax1.plot(Recall.round(2), color="r", label ='Recall')
#ax1.plot(sampleSize, ROCAUC.round(2), color="b", label ='ROC AUC')
ax1.tick_params(axis='y')
#ax1.xticks(5000)
#ax1.set_xscale("linear")
#ax1.set_xticks(sampleSize)
#ax1.set_xlim(50000, 0)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('Specificity')  # we already handled the x-label with ax1
ax2.plot(Specificity.round(2), color="b", label ='Specificity')
ax2.tick_params(axis='y')

ax1.legend()
ax2.legend()
#plt.legend()
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
