The result file which have been saved during the experiment execution if structured as follow
**XBrut**(n-1 columns) **YBrut** (1 Column from the dataset) **YClassified** (1 Column from the execution) **scores** (1 column) **pathLength** (1 column)

The function used to save the results data is "save_execution_results_data" defined in performances.py
**save_execution_results_data(self, YClassified,
                            scores, pathLengths, 
                            data_name = "Data", method_name="Method", 
                            execution_object="Experiment", 
                            alldataset = None, 
                            XBrut = None, YBrut = None)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
import sys
sys.path.append('../../../../../../../../')
from datasets import datasets as datat
dts = datat.synthetic_datasets(path_top="../../../../../../../../")

util = datat.utilitaries()

from metrics import useful_infos as ue

#from experiments import trainning_parametors
#t_parametors = trainning_parametors.functions_train_parametors()

image_folder = "saved_images/"
```

```python
def show_figure(data_path, dataset_name, method_name) :
    # Load the dataset
    dataset = datat.load_data_from_CSV(data_path);
    # Split in execution results
    XBrut, YBrut, YClassified, Scores, PathLengths = util.split_in_results(dataset)
    #print(XBrut)
    #print(PathLengths)
    #print(Scores)
    #print(YClassified)
    #print(YBrut)
    #plt.scatter(XBrut[0], XBrut[1], c="b")
    #plt.xlim(-30,30)
    #plt.ylim(-20,20)
    #plt.xlabel('X')
    #plt.ylabel('Y')
    #plt.title(method_name+" "+dataset_name+" Classified normal")
    #plt.show()
     
    #return
    # Concat in order to put the YClassified at the end and get Classified normal and classified abnormal
    # Important : In this case splitOn="PredictClass" to split on the classification
    X_normal, X_abnormal, pdataX = util.concat_columns_split(dataX=XBrut, dataScores=Scores, dataY=YBrut, 
                             outlier_label=ue._OUTLIER_PREDICTION_LABEL, 
                             dataYPrediction=YClassified, dataPathLength=PathLengths, splitOn="PredictClass")
    
    # Plot classified normal data
    plt.scatter(X_normal[0], X_normal[1], c="b")
    plt.xlim(-30,30)
    plt.ylim(-20,20)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(method_name+" "+dataset_name+" Classified normal")
    # Save picture
    plt.savefig(image_folder+method_name+"_"+dataset_name+"classiedNormal.pdf", bbox_inches='tight', pad_inches=0.05)
    
    plt.show()
    
    # Plot classified abnormal data
    plt.scatter(X_abnormal[0], X_abnormal[1], c="b")
    plt.xlim(-30,30)
    plt.ylim(-20,20)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(method_name+" "+dataset_name+" Classified abnormal")
    # Save picture
    plt.savefig(image_folder+method_name+"_"+dataset_name+"classifiedAbnormal.pdf", bbox_inches='tight', pad_inches=0.05)
    
    plt.show()
```
