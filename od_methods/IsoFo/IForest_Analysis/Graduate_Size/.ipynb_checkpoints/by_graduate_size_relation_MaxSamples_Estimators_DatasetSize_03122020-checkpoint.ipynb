## Transformation de la classe à expliquer

L'objectif ici est de remplacer les "o" par -1 et les "n" par 1 afin de faire les matrices de confusion avec la fonction dédiée de scikit-learn.

```python
y_transform_KDD99CUP      = y_brut_KDD99CUP
y_transform_KDD99CUP      = np.where(y_transform_KDD99CUP =='o',-1,1)
#y_transform_KDD99CUP

y_transform_sub_dataset_6 = np.where(y_sub_dataset_6 =='o',-1,1)
y_transform_sub_dataset_5 = np.where(y_sub_dataset_5 =='o',-1,1)
y_transform_sub_dataset_4 = np.where(y_sub_dataset_4 =='o',-1,1)
y_transform_sub_dataset_3 = np.where(y_sub_dataset_3 =='o',-1,1)
y_transform_sub_dataset_2 = np.where(y_sub_dataset_2 =='o',-1,1)
y_transform_sub_dataset_1 = np.where(y_sub_dataset_1 =='o',-1,1)
```

```python
X_recomposed_shuttle = dict()
X_recomposed_shuttle[0] = X_sub_dataset_1
X_recomposed_shuttle[1] = X_sub_dataset_2
X_recomposed_shuttle[2] = X_sub_dataset_3
X_recomposed_shuttle[3] = X_sub_dataset_4
X_recomposed_shuttle[4] = X_sub_dataset_5
X_recomposed_shuttle[5] = X_sub_dataset_6

Y_recomposed_shuttle = dict()
Y_recomposed_shuttle[0] = y_transform_sub_dataset_1
Y_recomposed_shuttle[1] = y_transform_sub_dataset_2
Y_recomposed_shuttle[2] = y_transform_sub_dataset_3
Y_recomposed_shuttle[3] = y_transform_sub_dataset_4
Y_recomposed_shuttle[4] = y_transform_sub_dataset_5
Y_recomposed_shuttle[5] = y_transform_sub_dataset_6
print(X_recomposed_shuttle)
print("********************************************************************")
print(Y_recomposed_shuttle)
```

```python
MS_executions_time = []
MS_roc_auc = []
MS_precisions = []
MS_recalls = []
MS_f1_scores = []
MS_tn = []
MS_fp = []
MS_fn = []
MS_tp = []
MS_specificity = []
MS_max_samples = []
MS_dataset_size = []
MS_estimators = []

min_roc_auc = 0.999

X_data = X_recomposed_shuttle[0]
Y_data = Y_recomposed_shuttle[0]

for j in range(0, len(X_recomposed_shuttle), 1):
    print("*************************************************************************************************")
    if(j > 0):
        #X_data = X_data.append(X_recomposed_shuttle[j], ignore_index=True, sort=False)
        #Y_data = Y_data.append(Y_recomposed_shuttle[j], ignore_index=True, sort=False)
        X_data = np.concatenate((X_data , X_recomposed_shuttle[j]))
        Y_data = np.concatenate((Y_data , Y_recomposed_shuttle[j]))
    MS_dataset_size.append(len(X_data))
    print("Dataset size : ")
    print(len(X_data))
    
    intermediate_roc_auc = []
    intermediate_max_samples = []
    intermediate_execution_time = []
    
    found = False
    z = 0
    while found == False and z < len(X_data):
        #for z in range(100, len(X_data), 100):
        z = z + 100
        if z > len(X_data):
            z = len(X_data)
        max_samples = z
        
        found_estimators = False
        n_estimators = 0
        while found_estimators == False and n_estimators <= 200:
            n_estimators = n_estimators + 10
            start = time()

            # Instanciation, fit and predict on Shuttle
            MS_func = IsolationForest(behaviour="new", max_samples=max_samples, n_estimators=n_estimators)
            MS_func.fit(X_data)
            MS_y_pred = MS_func.predict(X_data)

            # Calcul du temps d'exécution
            MS_exec_time = time() - start

            # Calcul de l'aire sous la courbe ROC
            MS_y_DF = MS_func.decision_function(X_data)
            MS_auc = roc_auc_score(Y_data, MS_y_DF)

            if MS_auc >= min_roc_auc:
                found = True
                found_estimators = True

                MS_estimators.append(n_estimators)
                print("N estimators: ")
                print(n_estimators)

                MS_max_samples.append(max_samples)
                print("Max samples: ")
                print(max_samples)

                MS_executions_time.append(MS_exec_time)
                print("Temps d'exécution: ")
                print(MS_exec_time)

                MS_roc_auc.append(MS_auc)
                print("ROC AUC: ")
                print(MS_auc)

                # Precision
                MS_precision = precision_score(Y_data, MS_y_pred)
                MS_precisions.append(MS_precision)
                print("Precision: ")
                print(MS_precision)

                # Recall
                MS_recall = recall_score(Y_data, MS_y_pred)
                MS_recalls.append(MS_recall)
                print("Recall: ")
                print(MS_recall)

                # f1_score
                MS_f1_score = f1_score(Y_data, MS_y_pred)
                MS_f1_scores.append(MS_f1_score)
                print("f1_score: ")
                print(MS_f1_score)

                # Specificity
                MS_specificit = specificity_score(Y_data, MS_y_pred)
                MS_specificity.append(MS_specificit)
                print("Specificity: ")
                print(MS_specificit)

                # Matrice de confusion
                print("Matrice de confusion: ")
                print(confusion_matrix(Y_data, MS_y_pred))
                ttn, tfp, tfn, ttp = confusion_matrix(Y_data, MS_y_pred).ravel()
                tn = ttp
                fp = tfn
                fn = tfp
                tp = ttn
                MS_tn.append(tn)
                MS_fp.append(fp)
                MS_fn.append(fn)
                MS_tp.append(tp)
```
