```python
# Remind : Raja toolk about scikit learn french book. 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
shuttle_data = pd.read_csv('/Users/thesard/Doctorat/These2018/ISEP/Developpements/EspaceIntelliJ/LearningAllInPython/Learning_IHM_Features/datasets/imported_datasets/shuttle-unsupervised-ad_2019-06-14 17:33:02.493755.csv', header=None, index_col=None)
shuttle_data_used = shuttle_data[[0, 1, 2, 3, 4, 5, 6, 7, 8]]
print(shuttle_data.all)
print("Kurtosis result : {}".format(shuttle_data_used.kurtosis()))
print("Skew result : {}".format(shuttle_data_used.skew()))
```

```python
n_outliers = shuttle_data[shuttle_data[9] == 'o']
n_inliers = shuttle_data[shuttle_data[9] == 'n']
print("Outliers in the dataset = \n {}".format(n_outliers))
print("Inliers in dataset = \n {}".format(n_inliers))
```

```python
from sklearn.model_selection import train_test_split
#X_train, X_test, y_train, y_test = train_test_split(shuttle_data_used, test_size=0.33, random_state=42)
X_train, X_test = train_test_split(shuttle_data, test_size=0.33, random_state=42)
print("X_train = \n {}".format(X_train.all))
print("X_test = \n {}".format(X_test.all))
#print("y_train = \n {}".format(y_train.all))
#print("y_test = \n {}".format(y_test.all))
```

```python
n_outliers_train = X_train[X_train[9] == 'o']
n_outliers_test = X_test[X_test[9] == 'o']
n_outliers_train_size = len(n_outliers_train)
n_outliers_test_size = len(n_outliers_test)
print("Outliers in the X_train = \n {}".format(n_outliers_train))
print("Inliers in the X_test = \n {}".format(n_outliers_test))
```

```python
X_train_used = X_train[[0, 1, 2, 3, 4, 5, 6, 7, 8]]
X_test_used = X_test[[0, 1, 2, 3, 4, 5, 6, 7, 8]]
```

```python
from sklearn import decomposition
pca_train = decomposition.PCA(n_components=2)
pca_train.fit(X_train_used)
X_train_used_2 = pca_train.transform(X_train_used)
print("Result of PCA on X_train_used_2 = {}".format(X_train_used_2))
print("Result of PCA on X_train_used_2 = {}".format(X_train_used_2.shape))

pca_test = decomposition.PCA(n_components=2)
pca_test.fit(X_test_used)
X_test_used_2 = pca_test.transform(X_test_used)
print("Result of PCA on X_test_used_2 = {}".format(X_test_used_2))
print("Result of PCA on X_test_used_2 = {}".format(X_test_used_2.shape))
```

```python
from sklearn.svm import OneClassSVM
nu = 0.05  
gamma = 0.05 
ocsvm = OneClassSVM(kernel='rbf', gamma=gamma, nu=nu)
ocsvm.fit(X_train_used_2)
```

```python
y_pred_train = ocsvm.predict(X_train_used_2)
y_pred_test = ocsvm.predict(X_test_used_2)
print("Result of prediction on X_train_used_2 = {}".format(y_pred_train))
print("Result of prediction on X_test_used_2 = {}".format(y_pred_test))
```

```python
n_error_train = y_pred_train[y_pred_train == -1].size
n_error_test = y_pred_test[y_pred_test == -1].size
print("Number of outliers found in X_train_used_2 = {}".format(n_error_train))
print("Number of outliers found in X_test_used_2 = {}".format(n_error_test))
# The problem here is how to match the PCA result with the inputed dataset for more understanding
# Try to read and understand Jade sended document about PCA
```

```python
import matplotlib.font_manager
s = 40
b1 = plt.scatter(X_train_used_2[:, 0], X_train_used_2[:, 1], c='white', s=s, edgecolors='k')
b2 = plt.scatter(X_test_used_2[:, 0], X_test_used_2[:, 1], c='blueviolet', s=s,edgecolors='k')
plt.axis('tight')
plt.legend([b1, b2],
           ["training observations",
            "test observations"],
           loc="upper left",
           prop=matplotlib.font_manager.FontProperties(size=11))
plt.xlabel(
    "Outliers in train: %d/%d ; Outliers in test: %d/%d ; "
    % (n_error_train, n_outliers_train_size, n_error_test, n_outliers_test_size))
plt.show()
```
